```python
# imports
import pandas as pd
```

```python
# read in the data
data = pd.read_csv('Datasets/calcofi_combined.csv')
data.head()
```

```python
# sort the data
data_sorted = data.sort_values(by=['Julian_Date'], ascending = True)

# Check the sorted data
data_sorted[['Sta_ID', 'Julian_Date', 'ChlorA']]
```

```python
# data_chla = data_sorted.pivot_table(index='Julian_Date', columns='Sta_ID', values='ChlorA')
data_cla = data_sorted.pivot_table(index='Julian_Date', columns='Sta_ID', values='ChlorA')
data_chla.head()
data_chla.to_csv('Datasets/data_SSA_timeseries.csv', index=False)
```

```python
# Filter columns that have more than 150 non-null values
filtered_data = data_chla.loc[:, data_chla.notnull().sum(axis=0) > 150]
filtered_data
```
